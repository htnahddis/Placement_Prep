```python
from litellm import completion

response = completion(
    model="gemini/gemini-2.5-flash",
    messages=[
        {"role": "user", "content": "Write a short story on AI."}
    ],
    stream=True
)

for chunk in response:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end="")
```

```python
from litellm import completion

def ask_llm(model, prompt):
    response = completion(
        model=model,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    return response.choices[0].message.content


print(ask_llm(
    "gemini/gemini-2.5-flash",
    "What is RAG?"
))

print()

print(ask_llm(
    "groq/llama-3.3-70b-versatile",
    "What is RAG?"
))
```

```python
from litellm import completion

prompt = "Explain embeddings in simple terms."

models = [
    "gemini/gemini-2.5-flash",
    "groq/llama-3.3-70b-versatile"
]

for model in models:
    response = completion(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    print("=" * 60)
    print(model)
    print(response.choices[0].message.content)
```

```python
from litellm import completion
import time

models = [
    "gemini/gemini-2.5-flash",
    "groq/llama-3.3-70b-versatile"
]

prompt = "Explain retrieval augmented generation."

for model in models:

    start = time.time()

    response = completion(
        model=model,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    elapsed = time.time() - start

    print("=" * 60)
    print("Model:", model)
    print("Time:", round(elapsed, 2), "seconds")
    print(response.choices[0].message.content)
```
