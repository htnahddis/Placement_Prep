```python
#create simple txt file 
import os 

os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_text = {
    "../data/file1.txt": """
    This is sample content for file 1.
    It contains around 100 words of text.
    You can use this content for testing file reading,
    vector databases, embeddings, document loaders,
    or any NLP-related task. The content discusses
    artificial intelligence, machine learning,
    software development, and data processing.
    Engineers often work with large datasets and
    require efficient methods to store, retrieve,
    and analyze information. Python provides many
    useful libraries such as pandas, numpy, and
    scikit-learn that simplify these tasks.
    """,

    "../data/file2.txt": """
    This is sample content for file 2.
    The document focuses on web development using
    modern frameworks. React is a popular JavaScript
    library for building user interfaces, while
    backend services can be created using Flask,
    Django, or FastAPI. Developers frequently use
    Git for version control and GitHub for
    collaboration. Good software design emphasizes
    maintainability, readability, and scalability.
    Testing and continuous integration help ensure
    that applications remain reliable as they grow.
    """
}

for filepath, content in sample_text.items():
    with open(filepath, 'w', encoding="utf-8") as f:
        f.write(content)


print("Sample files completed!")
```

```python
##text loader 
# from langchain.document_loaders import TextLoader - deprecated
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/file1.txt", encoding= "utf-8")
document = loader.load()
print(document)
```

```python
## Directory Loader 
from langchain_community.document_loaders import DirectoryLoader

## load all the text files from the directory
dir_loader = DirectoryLoader(
    "../data/text_files",
    glob = "**/*.txt",
    loader_cls = TextLoader, 
    loader_kwargs = {"encoding": "utf-8"},
    show_progress= False

)

doc = dir_loader.load()
doc
```
