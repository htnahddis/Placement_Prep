```python


import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv('GEMINI_API_KEY')
if api_key is None:
    raise RuntimeError('GEMINI_API_KEY is not set. Please add it to your .env file or environment variables.')
```

## Basic creating of agent

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.tools import tool
from langchain.agents import create_agent

def get_weather(location: str) -> str:
    """This is a mock implementation. Replace with actual API call to a weather service."""
    return f"The current weather in {location} is sunny with a temperature of 25°C."


gemini = ChatGoogleGenerativeAI(api_key=api_key, model="gemini-2.5-flash")

agent = create_agent(
    model=gemini,
    tools=[get_weather],

)

agent

agent.invoke({"messages":[{"role": "user", "content": "What is the weather in France?"}]})
```
