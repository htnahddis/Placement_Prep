```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['Aman', 'Riya', 'Kabir', 'Sneha', 'Aryan', 'Meera'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT'],
    'Marks': [85, 72, np.nan, 90, 66, 78],
    'Age': [21, 22, 20, 23, np.nan, 21]
})
```

```python
display(df[:4])
display(df[3:])
```

```python
df.isnull().count()
```

```python
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)
display(df)
```

```python
df[df['Marks'] > 75]
```

```python
df['Status'] = df['Marks'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
display(df)
```

```python
df.groupby('Department')['Marks'].mean()
```

```python
df.sort_values(by='Marks', ascending=False, inplace=True)
display(df)
```
