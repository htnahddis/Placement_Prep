```python
import pandas as pd
import numpy as np

orders = pd.DataFrame({
    'OrderID': [101,102,103,104,105,106,107,108,109,110],
    'Customer': ['Aman','Riya','Kabir','Aman','Sneha','Riya','Kabir','Meera','Aman','Sneha'],
    'Category': ['Electronics','Fashion','Electronics','Groceries','Fashion',
                 'Electronics','Fashion','Groceries','Electronics','Fashion'],
    'Amount': [5000,2000,np.nan,1500,3000,7000,2500,1200,5000,3000],
    'City': ['Mumbai','Delhi','Mumbai','Pune','Delhi',
             'Mumbai','Pune','Mumbai','Delhi','Delhi'],
    'OrderDate': ['2025-01-10','2025-01-12','2025-01-15','2025-02-01',
                  '2025-02-10','2025-02-18','2025-03-01',
                  '2025-03-05','2025-03-20','2025-03-25']
})

customers = pd.DataFrame({
    'Customer': ['Aman','Riya','Kabir','Sneha','Meera'],
    'Age': [21,24,22,23,25],
    'Membership': ['Gold','Silver','Gold','Platinum','Silver']
})
```

```python
# Task 1 — Basic Inspection

# Display:

# first 5 rows
# dataframe shape
# column names
# data types

display(orders.head(5))
display(orders.shape)
display(orders.columns)
display(orders.dtypes)

```

```python
display(orders.isnull())
orders.fillna(orders['Amount'].mean(), inplace = True)
display(orders.isnull())
```

```python
orders.OrderDate = pd.to_datetime(orders.OrderDate)
display(orders.dtypes)

#give  month names in month column

orders['Month'] = orders.OrderDate.dt.month_name()
display(orders.head(5))
```

```python
display(orders[orders['Amount'] > 3000])
display(orders[(orders['City'] == 'Mumbai') & (orders['Category'] == 'Electronics')])
display(orders[(orders['Category'] == 'Groceries') | (orders['Category'] == 'Fashion')])
```

```python
orders['Customer'] = orders['Customer'].str.upper()
display(orders.head(5))
```

```python
orders.to_csv('orders.csv', index = False)
```
